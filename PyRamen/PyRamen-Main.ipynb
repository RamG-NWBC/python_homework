```python
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter Modified by Ram Gollakota"""
# -*- coding: UTF-8 -*-
"""PyRamen Homework Starter Modified by Ram Gollakota"""

# @TODO: Import libraries
import csv
from pathlib import Path

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('Resources/menu_data.csv')
sales_filepath = Path('Resources/sales_data.csv')

# Print file paths to verify access to files
print(f'menu_filepath: {menu_filepath}')
print(f'sales_filepath: {sales_filepath}')

# @TODO: Initialize list objects to hold our menu and sales data
menu_list = []
sales_list = []



# @TODO: Read in the menu data into the menu list

# Open the menu csv file as an object
with open(menu_filepath, 'r') as menu_file:
  # Create a reader object
    reader_menu = csv.reader(menu_file)
    #skip the header row
    next(reader_menu)
  # Iterate over the rows
    for row in (reader_menu):
    # append to sales
        menu_list.append(row)
    print(menu_list[:10])

print(" now the Sales data")

# @TODO: Read in the sales data into the sales list

with open(sales_filepath, 'r') as sales_file:
  # Create a reader object
    reader_sales = csv.reader(sales_file)
    #skip the header row
    next(reader_sales)
  # Iterate over the rows
    for row in (reader_sales):
    # append to sales
        sales_list.append(row)
    print(sales_list[:10])
    
   

# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Initialize a row counter variable
row_count = 0

# @TODO: Loop over every item in the sales list object

for row in sales_list:
    line_item_id = row[0]
    date = row[1]
    credit_card_number = row[2]
    quantity = row[3]
    menu_item = row[4]
    quantity = int(quantity)

   
    # Get the price and cost for the menu item from the menu list
    for item_row in menu_list:
        item = item_row[0]
        price = item_row[3]
        cost = item_row[4]
        price = float(price)
        cost = float(cost)
        
        
        # if item name is in any of the sales then start tracking it
        if item == menu_item:
            
            #print(f"Found menu item: {menu_item}")
            # if item is not found in report the add metrics for each item 
            
            if item not in report:
                report[item] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0}
            #increment count for item - as item is now found in report
            report[item]["01-count"] += quantity
            report[item]["02-revenue"] += price * quantity
            report[item]["03-cogs"] += cost * quantity
            report[item]["04-profit"] += (price - cost) * quantity
        
        
   
      # Increment the row counter
    row_count += 1
    
    import csv

# Open a file in write mode
with open('report.csv', 'w', newline='') as csvfile:
  # Create a writer object
  writer = csv.writer(csvfile)
  # Write the header row
  writer.writerow(["Item", "Count", "Revenue", "Cost", "Profit"])

  # Loop over the items in the report dictionary
  for item, metrics in report.items():
    # Write a row for each item
    writer.writerow([item, metrics["01-count"], metrics["02-revenue"], metrics["03-cogs"], metrics["04-profit"]])

```
